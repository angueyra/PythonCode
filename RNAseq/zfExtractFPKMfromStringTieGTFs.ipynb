```python
def extractTranscripts(fName):
    dPath = "~/Documents/LiMolec/zfRNAseq/20190827/20190827_DE/" #directory path
    outPath = "Transcripts/"
    # insert header & create file
    !echo $'seqname\tsource\tstart\tend\tscore\tstrand\tframe\tgene_id\ttranscript_id\tcov\tFPKM\tTPM' > {dPath + outPath + fName}
#     # extract only trancripts (and ignore exon information) and append to file
#     # EXACT MATCH ONLY by matching word (-w)
    !grep -w transcript {dPath + fName} >> {dPath + outPath + fName}
#     # transform attributes into columns
    !sed -i '' $'s/transcript\t//' {dPath + outPath + fName}
    !sed -i '' 's/gene_id "//' {dPath + outPath + fName}
    !sed -i '' $'s/"; transcript_id "/\t/' {dPath + outPath + fName}
    !sed -i '' $'s/"; cov "/\t/' {dPath + outPath + fName}
    !sed -i '' $'s/"; FPKM "/\t/' {dPath + outPath + fName}
    !sed -i '' $'s/"; TPM "/\t/' {dPath + outPath + fName}
    !sed -i '' $'s/";//' {dPath + outPath + fName}
    
    
    
fNames = [
    '01_zfL1.gtf',
    '02_zfL2.gtf',
    '03_zfL3.gtf',
    '04_zfU5.gtf',
    '05_zfM1.gtf',
    '06_zfM2.gtf',
    '07_zfM3.gtf',
    '08_zfR1.gtf',
    '09_zfR2.gtf',
    '10_zfR3.gtf',
    '11_zfR4.gtf',
    '12_zfL4.gtf',
    '13_zfL5.gtf',
    '14_zfL6.gtf',
    '15_zfS5.gtf',
    '16_zfS6.gtf',
    '17_zfM4.gtf',
    '18_zfM5.gtf',
    '19_zfM6.gtf',
    '20_zfS7.gtf',
    '21_zfR5.gtf',
    '22_zfR6.gtf',
    '23_zfU1.gtf',
    '24_zfU2.gtf',
    '25_zfU3.gtf',
    '26_zfU4.gtf',
    '27_zfS1.gtf',
    '28_zfS2.gtf',
    '29_zfS3.gtf',
    '30_zfS4.gtf',
];
      
### Run only once
# for f in fNames:
#      print(f)
#      extractTranscripts(f)
# print('Listo!')
```

### Load gtf file for a particular sample and fill out column

```python
dPath = "~/Documents/LiMolec/zfRNAseq/20190827/20190827_DE/Transcripts/" #directory path

fNames = [
    'L1',
    'L2',
    'L3',
    'L4',
    'L5',
    'L6',
    'M1',
    'M2',
    'M3',
    'M4',
    'M5',
    'M6',
    'R1',
    'R2',
    'R3',
    'R4',
    'R5',
    'R6',
    'S1',
    'S2',
    'S3',
    'S4',
    'S5',
    'S6',
    'S7',
    'U1',
    'U2',
    'U3',
    'U4',
    'U5',
]

def getFPKM(genename,df):
    return df['FPKM'][df['gene_id'] == genename].sum()

```

```python
for fN in fNames:
    sampleGDF = pd.read_csv(dPath + fN + ".gtf", delimiter='\t')
    print ('Loaded ' + fN)
    for i in range(gdf.shape[0]):
        gdf.loc[i,fN]=getFPKM(gdf['symbol'][i],sampleGDF)
    print ('Listo')
# gdf.head(10)
```
