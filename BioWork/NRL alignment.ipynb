```python
from Bio import Entrez
from Bio import SeqIO
#Entrez.efetch?
```

```python
Entrez.email='angueyra@nih.gov'
def fetchgene(gbid,filename):
    temp = Entrez.efetch(db="protein", rettype="gb", id=gbid)
    gbseq = SeqIO.read(temp, "gb") #gb - genebank file format
    #open an output file in the current directory and write data to it
    out = open("data/"+filename+".gbk",'w')
    # To download to a specific location using an absolute path try something like this:
    #out = open("/Users/squiresrb/Documents/BCBB/Seminars/2015/Introduction to Biopython/ebola.gbk",'w')
    SeqIO.write(gbseq, out, "gb")
    temp.close()
    out.close()
    SeqIO.convert("data/"+filename+".gbk", "gb","data/"+filename+".fa", "fasta")
```

```python
fetchgene("NP_001129546.1","m_NRL")
fetchgene("XP_005338859.1","sq_NRL")
fetchgene("NP_006168.1","h_NRL")
```

```python
mNRL = SeqIO.read(open("data/m_NRL.fa"), format="fasta")
sqNRL = SeqIO.read(open("data/sq_NRL.fa"), format="fasta")
hNRL = SeqIO.read(open("data/h_NRL.fa"), format="fasta")
print ("mouse NRL = "+ str(len(mNRL.seq)) + " aa")
print ("squirrel NRL = "+ str(len(sqNRL.seq)) + " aa")
print ("human NRL = "+ str(len(hNRL.seq)) + " aa")
```

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
m2sq=pairwise2.align.globalxx(mNRL.seq,sqNRL.seq)
```

```python
print(format_alignment(*m2sq[0]))
```

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
h2sq = pairwise2.align.globalxx(hNRL.seq,sqNRL.seq)
```

```python
print(format_alignment(*h2sq[0]))
```
