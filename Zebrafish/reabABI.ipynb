```python
from Bio import SeqIO
import matplotlib
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python
abi_path = '/Users/angueyraaristjm/Documents/LiMolec/zfGenotyping/20190304_nrxn3a_larvae/ReSeqResults/'
abi_files = !ls $abi_path
abi_i = 8
abi_files[abi_i]
abi_path + abi_files[abi_i]
```

```python
abi_data = SeqIO.read(abi_path + abi_files[abi_i], 'abi')

# Base corresponding to DATA9 - 12
baseorder = abi_data.annotations['abif_raw']['FWO_1']
# Chromatogram data for each channel
channels = ['DATA9', 'DATA10', 'DATA11', 'DATA12']
chromatograms = defaultdict(list)

for i in range(0,4):
    chromatograms[baseorder[i]] = abi_data.annotations['abif_raw'][channels[i]]

# G - black, A - green, C - blue, T - red
```

```python
plt.plot(chromatograms['G'], color='black')
plt.plot(chromatograms['A'], color='green')
plt.plot(chromatograms['C'], color='blue')
plt.plot(chromatograms['T'], color='red')
plt.xlim((450, 650))
# plt.ylim((0, 2000))
plt.show()
```
